```python
import requests
import json

data_id = 897
api_key = '6c1fee8ca4cb9fed37f0a7047d2a5ad5'
data_url = f'https://apidata.mos.ru/v1/datasets/{data_id}/rows?api_key={api_key}'
data_get = requests.get(data_url)

print(data_get)
```

```python
bicycle_data = data_get.json()
print(bicycle_data[1])

out_list = list()

for value in bicycle_data:
  if value['Cells']['OperOrgName'] == 'ГБУ Автомобильные дороги ЦАО':
    out_list.append(value['Cells']['Width'])

print(out_list)
```

```python
# Проверяю данные в виде ширин полос (м) по наименованию эксплуатирующей организации «ГБУ Автомобильные дороги ЦАО»
import matplotlib.pyplot as plt
import numpy as np


x = [i for i in range(len(out_list))]
plt.scatter(x, out_list)

y = np.polyfit(x, out_list, 1)
p = np.poly1d(y)

plt.plot(x, p(x), 'deepskyblue')
plt.show()
```

```python
y_gradient = np.gradient(out_list)

y = np.polyfit(x, y_gradient, 1)
p = np.poly1d(y)

plt.scatter(x, y_gradient)
plt.plot(x, p(x), 'orchid')
plt.show()
```
